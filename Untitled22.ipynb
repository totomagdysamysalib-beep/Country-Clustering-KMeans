```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Country-data.csv')
dataset.set_index('country',inplace=True)
x=dataset.values
```

```python
dataset.isnull().sum()
```

```python
print(dataset.duplicated().sum())
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmean=KMeans(n_clusters=i,init="k-means++",random_state=42)
  kmean.fit(x)
  wcss.append(kmean.inertia_)
plt.plot(range(1,11),wcss)
```

```python
kmean=KMeans(n_clusters=3,init="k-means++",random_state=42)
y_kmean=kmean.fit_predict(x)
print(y_kmean)
```

```python
dataset["cluster_id"]=y_kmean
analysis=dataset.groupby("cluster_id")[["child_mort","income","inflation","gdpp"]].mean()
print(analysis)
```

```python
for country_id in range(3):
  countries=dataset[dataset["cluster_id"]==country_id].index.tolist()
  print(countries)
```
